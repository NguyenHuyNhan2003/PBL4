```python
from facebook_scraper import get_posts
import pandas as pd
```

```python
from datetime import datetime, timedelta

days_before = (datetime.now() - timedelta(days=7)).date() #date from which to scrap posts
print(days_before)
```

```python
reaction_count = 0;
post_list = []

for post in get_posts(FANPAGE_LINK,
                    options={"comments": True, "reactions": True, "allow_extra_requests": True},
                    extra_info=True, pages=2, cookies=COOKIE_PATH):
   if(post['time'].date() >= days_before): #limit the date from which to add to post_list
    if(post['reactions'] != None):
      if('like' in post['reactions']):
        like_count = post['reactions']['like']
        reaction_count += like_count
      else: like_count = 0
      if('love' in post['reactions']):
        love_count = post['reactions']['love']
        reaction_count += love_count
      else:  love_count = 0
      if('haha' in post['reactions']):
       haha_count = post['reactions']['haha']
       reaction_count += haha_count
      else: haha_count = 0
      if('sad' in post['reactions']):
        sad_count = post['reactions']['sad']
        reaction_count += sad_count
      else: sad_count = 0
      if('wow' in post['reactions']):
        wow_count = post['reactions']['wow']
        reaction_count += wow_count
      else: wow_count = 0
      if('angry' in post['reactions']):
        angry_count = post['reactions']['angry']
        reaction_count += angry_count
      else: angry_count = 0
      if('care' in post['reactions']):
        care_count = post['reactions']['care']
        reaction_count += care_count
      else: care_count = 0
    post_info = {
          "text": post["text"],
          "shares": post["shares"],
          "like_count": like_count,
          "love_count": love_count,
          "care_count": care_count,
          "haha_count": haha_count,
          "wow_count": wow_count,
          "sad_count": sad_count,
          "angry_count": angry_count,
          "reaction_count": reaction_count,
          "comments": post["comments"],
          "post_date": post["time"]
        }
    post_list.append(post_info)
```
